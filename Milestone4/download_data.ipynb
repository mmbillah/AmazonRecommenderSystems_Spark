## Unzip metadata files and save as csv

```python
import pandas as pd
import gzip
import json

for f in os.listdir():
    if not f.endswith('.gz'):
        continue
    elif f.split('.')[0] + '.csv' in os.listdir():
        continue
    def parse(path):
        g = gzip.open(path, 'rb')
        for l in g:
            yield json.loads(l)

    def getDF(path):
        i = 0
        df = {}
        for d in parse(path):
            df[i] = d
            i += 1
        return pd.DataFrame.from_dict(df, orient='index')

    df = getDF(f)
    df = df[['asin', 'title', 'image']].explode('image')
    df = df.groupby('asin').first().reset_index()

    df.to_csv(f.split('.')[0] + '.csv', index=False)
```
